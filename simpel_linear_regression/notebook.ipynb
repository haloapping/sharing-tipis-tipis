```python
# step 1: import modules
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# step 2: load data, X (feature) and y (target)
X, y = make_regression(n_samples=20,
                       n_features=1,
                       n_targets=1,
                       noise=10,
                       random_state=42)

# step 3: train
model = LinearRegression().fit(X, y)

# step 4: predict
y_pred = model.predict(X)

# step 5: evaluate
mae = mean_absolute_error(y, y_pred)

# step 6: visualize prediction
plt.figure(figsize=(5, 5))
plt.title(f"Simple Linear Regression (MAE: {mae:.3f})", size=13)
plt.xlabel("X (feature)")
plt.ylabel("y (target)")
plt.grid()
plt.scatter(X, y, c="green", label="Samples")
plt.plot(X, y_pred, c="gold", label="Prediction LinReg Model")
plt.legend()
plt.savefig("../simpel_linear_regression/simple_linreg_visual.jpg", dpi=200)
plt.show()
```
