```python
# step 1: import modules
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# step 2: load data
X, y = make_regression(n_samples=20,
                       n_features=1,
                       n_targets=1,
                       noise=10,
                       random_state=42)

# step 3: train
lin_reg_model = LinearRegression().fit(X, y)

# step 4: predict
y_preds = lin_reg_model.predict(X)

# step 4: evaluate
mae = mean_absolute_error(y, y_preds)

# step 4: visualize prediction
plt.figure(figsize=(5, 5))
plt.title(f"Simple Linear Regression (MAE: {mae:.3f})", size=13)
plt.xlabel("X")
plt.ylabel("y")
plt.grid()
plt.scatter(X, y, c="green", label="Samples")
plt.plot(X, y_preds, c="gold", label="Prediction LinReg Model")
plt.legend()
plt.savefig("../simpel_linear_regression/simple_linreg_visual.jpg", dpi=200)
plt.show()
```
